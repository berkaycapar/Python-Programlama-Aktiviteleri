### öncelikle earthquake.csv dosyamızı import ediyoruz

```sqlite3
.mode csv
.import earthquake.csv earthquake
```

### daha iyi görünmesi için header kısmının altına çizgi çektiriyoruz

```sqlite3
.mode columns
.headers on
```

#### Problem 1: earthquake.csv içerisinde sadece kocaeli ilindeki depremlerin bilgilerini gösterelim

```sqlite3
SELECT * FROM earthquake WHERE city="kocaeli";
```

#### Problem 2: Kocaeli'de 50km ve üzeri deprem derinliklerini gösterelim

```sqlite3
SELECT * FROM earthquake WHERE city="kocaeli" AND depth >= 50;
```

#### Problem 3: Sadece Kocaeli'deki 7 ve üzerindeki md+mw+ms+mb değerlerinin gösterilmesi

```sqlite3
SELECT city, xm, md+mw+ms+mb AS ŞİDDET_DOĞRULAMA FROM earthquake WHERE city="kocaeli" AND xm >= 7;
```

#### Problem 4: Kocaeli'deki depremlerin şiddetlerini gösteren tablo

```sqlite3
SELECT city, xm,
       CASE WHEN xm <3 THEN 'hafif şiddetli'
            WHEN xm<5 THEN 'orta şiddetli'
            WHEN xm<6 THEN 'çok şiddetli'
            WHEN xm>6 THEN 'ağır şiddetli'
            ELSE 'belirsiz' END
                AS şiddet_derecesi
    FROM earthquake
    WHERE city="kocaeli";
```

#### Problem 5: illere göre ortalama deprem şiddetleri (boş değerler olduğu için WHERE city!="" kullanıldı.)

```sqlite3
SELECT city, MAX(xm) FROM "earthquake" WHERE city!="" GROUP BY City
```

#### Problem 6: Türkiyenin kuzeyinde olan depremlerin ortalama derinlik değerleri ve illerin ismi

```sqlite3
SELECT city, AVG(depth), direction FROM "earthquake" GROUP BY City HAVING direction = "north"
```

#### Problem 7: Tüm iller için en yüksek deprem şiddeti değerleri ve depremin derinliği

```sqlite3
Select city, depth, MAX(xm) FROM "earthquake" GROUP BY City HAVING city!=""
```

#### Problem 8: Distinct kullanarak deprem olan illerin raporu

```sqlite3
SELECT DISTINCT city FROM earthquake where country="turkey" AND city != ""
```

#### Problem 9: ORDER BY ve ASC/DESC kullanarak deprem şiddetlerinin sıralanması.

```sqlite3
Select city, date, depth, xm FROM "earthquake" WHERE city!="" ORDER BY xm DESC
```

#### Problem 10: Intersect kullanarak xm>= 6 derinlik>= 50 depremleri bulmak

```sqlite3
Select city, date, depth, xm FROM "earthquake" WHERE xm >= 6 AND city!=""
INTERSECT
Select city, date, depth, xm FROM "earthquake" WHERE depth >= 50 and city!=""
```

#### Problem 11: Kocaelinde 01.01.1999 ile 31.12.2000 tarihleri arasında olan depremler, şiddetleri ve derinliği

```sqlite3
Select city, date, depth, xm FROM "earthquake" WHERE city="kocaeli"and date BETWEEN '1999.01.01' AND '2000.12.31'
```

#### Problem 12: Aydın'da 6.8 depremi hangi yıl oldu?

```sqlite3
Select city, date, xm FROM "earthquake" WHERE city="aydin" and xm = 6.8
```

#### Problem 13: Bölgelere göre deprem sayılarının istatistiği

```sqlite3
SELECT direction, count(xm) FROM "earthquake" GROUP BY direction HAVING direction!="" ORDER BY count(xm)DESC
```

#### Problem 14: richer ölçeği 0 olmayan deprem listesi

```sqlite3
SELECT country, city, date, xm, direction, richter FROM "earthquake" WHERE richter != 0 and country ="turkey" ORDER BY richter DESC
```

#### Problem 15: Türkiyedeki 10km derinlikte oluşan depremlerin ilçelerle birlikte gösterimi

```sqlite3
SELECT country, city, area, date, xm as şiddet, depth, richter FROM "earthquake" WHERE depth = 10 and country ="turkey"
```
