#### Problem 1:  Türkiye'deki 1970 ile 2005 arasındaki yıllara göre yaşanmış en büyük deprem büyüklüklerinin matplotlib.pyplot ile gösteriniz

```python
from datetime import datetime

fig = plt.figure(figsize = (20,15))
ax = fig.add_subplot(1, 1, 1)

data2 = pd.read_csv('earthquake.csv', usecols = ["date", "xm"],  index_col=0, parse_dates=True)
xm = data2['xm']

xm.plot(ax=ax, style='m+')

earthquake_data = [
    (datetime(1999, 8, 1), '99 Kocaeli Depremi'),
    (datetime(1976, 11, 24), '7.5 Büyüklüğündeki Van Depremi'),
    (datetime(1992, 3, 13), '92 Erzincan Depremi (6.8)')
]

for date, label in earthquake_data:
    ax.annotate(label, xy=(date, max(xm)),
                xytext=(date, max(xm)),
                arrowprops=dict(facecolor='blue', headwidth=14, width=12,
                                headlength=14))

ax.set_xlim(['1/1/1970', '1/1/2005'])
ax.set_ylim([3.5, 8.5])
ax.set_title('1970 ile 2005 arasındaki yıllara göre yaşanmış en büyük deprem büyüklüklerinin gösterimi ')
```

#### Problem 2: Aydın ve Kocaeli'nin 1910-2017 yılları arasındaki deprem sayılarını gösteren seaborn.barplot grafiğini çizdiriniz

```python
data = pd.read_csv('earthquake.csv')
plt.figure(figsize=(7,7))
ax=sns.barplot(x=data['city'].value_counts().index,
              y=data['city'].value_counts().values,
              order=["aydin", "kocaeli"])
plt.xlabel('Şehirler')
plt.ylabel('Deprem Sayilari')
plt.title('Aydın ve Kocaeli\'nin 1910-2017 yılları arasındaki deprem sayıları')
plt.show()
```

#### Problem 3: Depremin şiddeti ve derinliği arasında bir bağlantı var mıdır? İnceleyebileceğimiz bir grafik oluşturalım.

```python
sns.lmplot(x='xm',y='depth',data=data)
plt.xlabel('Depremin Şiddeti')
plt.ylabel('Derinliği')
plt.title('Depremlerin Şiddetlerine Göre Derinliklerinin Gösterimi ')
plt.show()
```

#### Problem 4: xm değerinin ms + richer + mb + md olduğunu biliyoruz. Peki bölgelere göre mb + md'nin diğer istatistiklere göre şekillenmesi nasıl olur?

```python
data['mbPLUSmd'] = (data['xm'] - data['ms'] - data['richter'])
data.head()
sns.barplot(x='mbPLUSmd', y='direction', data=data, orient='h')
```

#### Problem 5: Orta Karadeniz Bölgesindeki Deprem Şiddetlerinin Hangi Aralıklarda Yoğunlaştığını Araştıralım

```python
plt.figure(figsize=(720,720))
sns.jointplot(x=np.arange(1,577),y=data[(data['direction']=='north')].xm,color='red',space=0,kind='kde')
plt.ylabel('Deprem Şiddeti')
plt.title('Karadeniz Bölgesi Deprem Şiddet Yoğunluğu')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

#### Problem 6: Depremlerin Şiddet Değerlerinin (xm-md-richter-mw-ms-mb) Kutu Grafiği Kullanılarak Gösterilmesi Nasıl olur?

```python
NumOfQuake = data.drop(['id', 'date', 'time', 'lat', 'long',
                    'country', 'city', 'area', 'direction', 'mbPLUSmd',
                    'dist','depth'], axis = 1)
sns.boxplot(data = NumOfQuake)
sns.boxplot(data = NumOfQuake, notch = True, linewidth = 2.5, width = 0.50)

```

#### Problem 7: Deprem Şiddetlerinin sayılarını basit bir histogram grafiği ile gösterelim

```python
sns.distplot(data.xm)
```

#### Problem 8: Bölgelere göre deprem şiddetlerinin karşılaştırılmasını içeren bir Violin grafiği dağılımları kullanalım ve bunları kendi seçeceğimiz renk değerleri ile sembolize edelim.

```python
pkmn_type_colors = ['#78C850',  # north_west
                    '#F08030',  # south_west
                    '#6890F0',  # south_east
                    '#A8B820',  # north_east
                    '#A8A878',  # south
                    '#A040A0',  # north
                    '#F8D030',  # east
                    '#E0C068',  # west
                   ]

sns.violinplot(x=data['direction'], y='xm', data=data,
              palette = pkmn_type_colors )
```

#### Problem 9: Problem 8'deki sorunun aynısını bu sefer şiddetlere göre noktasal olarak swarm(sürü) grafiği belirtin. (Bu grafik bize her bir noktayı gösteriyor, ayrıca benzer noktaları birlikte gösteriyor)

```python
sns.swarmplot(x = data['direction'], y = 'xm', data = data, palette=pkmn_type_colors)
```
